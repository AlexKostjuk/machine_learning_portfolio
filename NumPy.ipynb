create filter mask

```python
f_cond = 1.5
s_cond = 5.
mask = ( a[:, 2] > f_cond) & (a[:, 0] < s_cond)
mask
```

change nan to 0.0 in array a

```python
a = np.nan_to_num(a, nan=0.0)
a
```

unique purpose and their number

```python
uniqs, counts =  np.array(np.unique(a, return_counts=True))

result_uniq = np.column_stack((uniqs, counts))
result_uniq
```

splitting an array into two parts

```python
result_split = np.vsplit(a, 2)
result_split
```

sort by increasing

```python
sorted_f_arr = result_split[0][np.argsort(result_split[0][:, 0])]
sorted_f_arr
```

sort by decreasing

```python
sorted_s_arr = result_split[1][np.argsort(result_split[1][:, 0])[::-1]]
sorted_s_arr
```

Array concatenation

```python
a_stack = np.vstack([sorted_f_arr, sorted_s_arr])
a_stack
```

the most repeated value, the probability that there are several of them is taken into account

```python
max_count = np.max(result_uniq[:, 1])
max_value_quantity = result_uniq[result_uniq[:, 1] == max_count]
max_value_quantity
```

```python
def meang_processing_fix(arr):
    arr = np.where(arr < np.mean(arr), arr * 2, arr / 4)
    return arr
```

```python
a_stack[:, 2] = meang_processing_fix(a_stack[:, 2]) 
a_stack
```
